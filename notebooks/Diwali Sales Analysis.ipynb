# Imports

```python
import numpy as np
import pandas as pd
%matplotlib inline
import seaborn as sns
```

# Load the dataset

```python
df = pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df.shape
```

# Insight of the data

```python
df.head(10)
```

# Info of all the columns in the dataset

```python
df.info()
```

# Data Cleaing ( Removed unneccessary columns with null values )

```python
df.drop(['Status','unnamed1'],inplace=True,axis=1)
```

```python
df.head(10)
```

# Checking columns with null values

```python
pd.isnull(df).sum()
```

```python
df.shape
```

# Drop rows with null values

```python
df.dropna(inplace=True)
```

```python
df.shape
```

# Change datatype of Amount column to INT

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df.info()
```

```python
df.head(10)
```

```python
df.rename(columns={'Cust_name':'Naam'})
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
df.columns
```

# Who shops more based on gender

```python
ax = sns.countplot(x='Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

**Women shops more than men**

# Who has spent more money based on gender

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data = sales_gen)
```

**Women has spent more money as compare to men**

# Which age group shops more

```python
ax = sns.countplot(x='Age Group',hue='Gender',data=df)
for bars in ax.containers:
    print(1)
    ax.bar_label(bars)
```

**People between the age of 26-35 shops the most specially women of that age group**

# Which age group spends more amount of money

```python
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

**26-35 age group has sepnt the most money**

```python
df.columns
```

# no. of orders from each state ( top 10 states )

```python
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State',y='Orders',data = sales_state)
```

**People from Uttar Pradesh has made the most orders**

# Who spent more money in each state

```python
sales_state_amt = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(ascending=False,by='Amount').head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State',y='Amount',data= sales_state_amt)
```

**People from Uttar Pradesh has spent the most amount of money**

# Who spent more money in each state wrt Gender

```python
sales_state_amt = df.groupby(['State','Gender'],as_index=False)['Amount'].sum().sort_values(ascending=False,by='Amount')
sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(x='State',y='Amount',data= sales_state_amt,hue='Gender')
```

**Women from Uttar Pradesh are the ones spending the most amount of money**

# Who shops more wrt marital status

```python
ax = sns.countplot(x='Marital_Status',data = df)
sns.set(rc={'figure.figsize':(6,5)})
for bar in ax.containers:
    ax.bar_label(bar)
```

**Unmarried people has spent more money**

# who spent more money wrt marital status

```python
sales_marital = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',data = sales_marital,hue='Gender')
```

***Unmarried women has spent the most money***

# People from which occupation has spent the most

```python
ax = sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(30,5)})
for bar in ax.containers:
    ax.bar_label(bar)
```

***People from the IT sector shops the most***

# People from which occupation has spent the most amount of money

```python
sales_occ = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(22,5)})
sns.barplot(x='Occupation',y='Amount',data=sales_occ)
```

***IT sector has the most purchasing power followed by healthcare and aviation***

# Most selling product

```python
ax = sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(30,5)})
for bar in ax.containers:
    ax.bar_label(bar)
```

The maximum amount is spent on ***clothing & apparel*** followed by ***food*** and ***electronics & gadgets***

```python
sales_pr = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)
sns.set(rc={'figure.figsize':(22,5)})
sns.barplot(x='Product_Category',y='Amount',data = sales_pr)
```

***Food*** items are the ones where the most money is spent followed by ***clothing & apparel***, ***electronics & gadgets*** and ***footwear and shoes***
